```python
airlines['Unnamed: 1']= pd.to_numeric(airlines['Unnamed: 1'],errors='coerce')
airlines['Unnamed: 2']= pd.to_numeric(airlines['Unnamed: 2'],errors='coerce')
airlines['Unnamed: 3']= pd.to_numeric(airlines['Unnamed: 3'],errors='coerce')
airlines['Unnamed: 4']= pd.to_numeric(airlines['Unnamed: 4'],errors='coerce')
airlines
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch 
z = linkage(airlines.iloc[:,1:], method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=0.,  # rotates the x axis labels
    #leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```
