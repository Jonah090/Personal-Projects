```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets  
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing 
```

```python
fraud = pd.read_csv("C:/Users/H P/Desktop/ANACONDA/DESCISION TREE/ASSIGNMENT/Fraud_check.csv")
fraud
```

```python
bins = [0,30000,100000]
group_names = ['Risky','Good']
```

```python
fraud['Taxable_Income'] = pd.cut(np.array(fraud.TaxIncome),bins,labels=group_names)
fraud
```

```python
fraud.head(50)
```

```python
label_encoder = preprocessing.LabelEncoder()
fraud['Urban']= label_encoder.fit_transform(fraud['Urban']) 
fraud['Undergrad']=label_encoder.fit_transform(fraud['Undergrad'])
fraud['Marital.Status']=label_encoder.fit_transform(fraud['Marital.Status'])
```

```python
fraud
```

```python
x=fraud.iloc[:,0:5]
```

```python
x 
pd.set_option("display.max_rows", None) 
x 
```

```python
y=fraud.iloc[:,6]
y
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=40) 
```

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(x_train,y_train) 
```

```python
tree.plot_tree(model);
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(model)(model, 
                   feature_names=['Undergrad','Marital.Status','City.Population','Work.Experience','Urban'],  
                   class_names=fraud.Taxable_Income,
                   filled=True)
fig.savefig("decistion_tree.png")
```

```python
model1 = DecisionTreeClassifier(criterion = 'entropy',max_depth=4)
model1.fit(x_train1,y_train1) 
```
